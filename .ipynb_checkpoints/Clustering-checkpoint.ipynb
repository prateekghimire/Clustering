```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as img
%matplotlib inline
```

```python
image=img.imread('test2.jpg')
```

```python
image
```

```python
plt.imshow(image)
```

```python
red=[]
blue=[]
green=[]
```

```python
for row in image:
    for temp_r,temp_g,temp_b in row:
        red.append(temp_r)
        green.append(temp_g)
        blue.append(temp_b)
```

```python
df=pd.DataFrame({'red':red,'blue':blue,'green':green})
```

```python
df.head()
```

```python
df.red.max()
```

```python
from sklearn.cluster import KMeans
```

```python
sse=[]
```

```python
for k in range(1,8):
    print(k)
    model=KMeans(n_clusters=k)
    model.fit(df[['red','blue','green']])
    sse.append(model.inertia_)
```

```python
k_value=np.arange(1,8)
```

```python
plt.plot(k_value,sse)
plt.show()
```

```python
#Using elbow method we take 3 as number of clusters
```

```python
model=KMeans(n_clusters=3)
```

```python
model.fit(df[['red','green','blue']])
```

```python
model.inertia_
```

```python
cluster_centers=model.cluster_centers_
```

```python
cluster_centers
```

```python
predicted_cluster=model.predict(df[['red','green','blue']])
```

```python
df['predicted_cluster']=predicted_cluster
```

```python
total_values=df.predicted_cluster.unique()
```

```python
total_values
```

```python
temp=0
selected=0
```

```python
for i in total_values:
    count=len(df[df.predicted_cluster==i])
    if count>temp:
        selected=i
    temp=count
```

```python
selected
```

```python
r=int(cluster_centers[0][0])
g=int(cluster_centers[0][1])
b=int(cluster_centers[0][2])
```

```python
plt.title('Most common color')
plt.imshow([[[r,g,b]]])
plt.axis('off')
```
